## Placeholder Step Renderers

These are placeholder implementations that will be replaced with full implementations
in future sessions.

Each step receives an `InteractionContext` with:
- `state`: Current workflow state dictionary
- `request`: FastHTML request object
- `session`: FastHTML session object
- `data`: Data from the step's `data_loader` function

```python
#| export
def render_selection_step(
    ctx: InteractionContext  # Interaction context with state and data
) -> Any:  # FastHTML component
    """Render Phase 1: Source Selection & Ordering step."""
    return Div(
        H2(
            "Phase 1: Source Selection & Ordering",
            cls=combine_classes(font_size._2xl, font_weight.bold, m.b(4))
        ),
        Div(
            Span("Placeholder", cls=str(font_weight.bold)),
            ": This step will allow selecting and ordering transcription sources.",
            cls=combine_classes(alert, alert_colors.info, m.b(4))
        ),
        P(
            "Features to implement:",
            cls=combine_classes(font_weight.semibold, m.b(2))
        ),
        Div(
            P("• Federated query across transcription plugin databases", cls=str(m.b(1))),
            P("• Multi-select source blocks from different plugins", cls=str(m.b(1))),
            P("• Drag-and-drop reordering for multi-chunk sources", cls=str(m.b(1))),
            P("• Preview of selected text blocks", cls=str(m.b(1))),
            cls=combine_classes(m.l(4), m.b(4))
        ),
        id=StructureDecompHtmlIds.SOURCE_SELECTOR
    )
```
